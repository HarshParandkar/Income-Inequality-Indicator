**MODEL BUILDING**

**LINEAR REGRESSION**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
def Regression(x_train,y_train,x_test,y_test):
  regressor=LinearRegression()
  regressor.fit(x_train,y_train)
  predictions=regressor.predict(x_test)
  mse=mean_squared_error(y_test,predictions)
  r2=r2_score(y_test,predictions)
  print("Mean Squared Error:",mse)
  print("R2 Score:",r2)

Regression(x_train,y_train,x_test,y_test)
```

Linear regression is a supervised machine learning algorithm used for predicting a continuous outcome variable based on one or more predictor variables.

**DECISION TREE**

```python
from sklearn.tree import DecisionTreeRegressor
def Decision(x_train,y_train,x_test,y_test):
  model = DecisionTreeRegressor(random_state=42)
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  mse = mean_squared_error(y_test, y_pred)
  r2 = r2_score(y_test, y_pred)
  print("Mean Squared Error:",mse)
  print("R^2 Score:",r2)

Decision(x_train,y_train,x_test,y_test)
```

 It creates a model in the form of a tree structure where each node represents a decision based on the value of a particular feature.

**RANDOM FOREST**

```python
from sklearn.ensemble import RandomForestRegressor
rdf=RandomForestRegressor()
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
def Random_Forest(x_train,x_test,y_train,y_test):
  rdf=RandomForestRegressor()
  random=rdf.fit(x_train,y_train)
  predict_random=random.predict(x_test)
  print("R2 Score:",r2_score(y_test,predict_random))
  print("Mean Squared Error",mean_squared_error(y_test,predict_random))
  print("Mean Absolute Error",mean_absolute_error(y_test,predict_random))

Random_Forest(x_train,x_test,y_train,y_test)
```

Random Forest builds several decision trees during training and merges their predictions to provide a more stable and accurate result.

**POLYNOMIAL REGRESSION**

```python
from sklearn.preprocessing import PolynomialFeatures
def Polynomial(x_train,y_train,x_test,y_test):
  poly=PolynomialFeatures(degree=2)
  X_poly=poly.fit_transform(x_train)
  model = DecisionTreeRegressor(random_state=42)
  model.fit(X_poly,y_train)
  x_test_poly=poly.transform(x_test)
  predictions=model.predict(x_test_poly)
  mse=mean_squared_error(y_test,predictions)
  print("Mean Squared Error:",mse)
  r2=r2_score(y_test,predictions)
  print("R2 Score:",r2)

Polynomial(x_train,y_train,x_test,y_test)
```

Polynomial regression is a form of linear regression in which the relationship between the independent variable x and the dependent variable y is modeled as an n-th degree polynomial.

**COMPARISON**

```python
def Comparision(x_train,x_test,y_train,y_test):
  print("LINEAR REGRESSION:")
  Regression(x_train,y_train,x_test,y_test)
  print()
  print("DECISION TREE REGRESSION")
  Decision(x_train,y_train,x_test,y_test)
  print()
  print("RANDOM FOREST REGRESSION")
  Random_Forest(x_train,x_test,y_train,y_test)
  print()
  print("POLYNOMIAL REGRESSION")
  Polynomial(x_train,y_train,x_test,y_test)
  print()


Comparision(x_train,x_test,y_train,y_test)
```

When choosing a regression model, it's important to compare different models to select the one that best fits data and provides the most accurate predictions. Here, we have provide a simple comparison of linear regression, polynomial regression,decisiontree and random forest regression.
